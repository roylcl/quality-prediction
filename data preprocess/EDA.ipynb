```python
import pandas as pd 

path = 'raw.csv'

dataframe = pd.read_csv(path)
```

```python
dataframe
```

```python
# calculate the total records and parameters of the dataset 
print("Number of records:", len(dataframe))
print("Number of parameters:", len(dataframe.columns))
```

```python
# calculate the records for each condition 
dataframe.groupby('Condition')['Condition'].count()
```

```python
# splitting the data into 3 dataframe to analyse on the missing data 
grouped = dataframe.groupby(dataframe.Condition)
```

```python
# splitting Normal1 as data1 
data1 = grouped.get_group("Normal1")
data1
```

```python
# find out how many null values are there in each parameter for Normal1  
data1.isnull().sum()
```

```python
# in Normal1, we uses interpolate method as we found that there are no missing data in the first and last row 
data1 = data1.interpolate(method='linear',limit_direction='forward')
```

```python
data1
```

```python
# splitting Condition1 as data2 
grouped = dataframe.groupby(dataframe.Condition)
data2 = grouped.get_group('Condition1')
data2
```

```python
# find out how many null values are there in each parameter for Condition1
data2.isnull().sum()
```

```python
# in Condition1, uses interpolate method with limit direction forward to fill up the null values 
data2 = data2.interpolate(method='linear',limit_direction='forward')
```

```python
data2
```

```python
# as our first row in Condition1 has a null value, we used backward fill to fill up 
data2 = data2.fillna(method='bfill',axis=0)
```

```python
data2
```

```python
# splitting Condition5 as data3 
grouped = dataframe.groupby(dataframe.Condition)
data3 = grouped.get_group('Condition5')
data3
```

```python
# find out how many null values are there in each parameter for Condition5 
data3.isnull().sum()
```

```python
# in Condition5, we uses interpolate method as we found that there are no missing data in the first and last row
data3 = data3.interpolate(method='linear',limit_direction='forward')
```

```python
data3
```

```python
# combining the 3 dataframes 
frames = [data1, data2, data3]
result = pd.concat(frames)
```

```python
# the filled dataframe of the dataset 
result
```

```python
# find out how many null values are there in each parameter for all conditions after filling up the null values 
result.isnull().sum()
```
